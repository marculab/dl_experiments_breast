```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
ta = [13,4,54,12,34]
va = [23,14,24,42,24]
x = np.array(list(range(len(ta)))) + 1
```

```python
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.plot(x,ta, x, va)
plt.legend(['train loss', 'val loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
```
